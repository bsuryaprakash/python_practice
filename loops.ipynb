```python
# simple code to loop through a list of numbers and print them

numbers = [1, 2, 3, 4, 5]
for number in numbers:
    print(number)
```

```python
# simple code to loop through a range of numbers and print them
for i in range(1, 6):
    print(i)
```

```python
# simple code to loop through a list of strings and print them
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# simple code to loop through a list of numbers and calculate their squares
numbers = [1, 2, 3, 4, 5]
squares = []
for number in numbers:
    squares.append(number ** 2)
print(squares)
```

```python
#simple code to loop through dictionary and print keys and values
person = {"name": "Alice", "age": 30, "city": "New York"}
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# simple code to show append, and break in a loop
numbers = [1, 2, 3, 4, 5]
squares = []
for number in numbers:
    if number == 4:
        break
    squares.append(number ** 2)
print(squares)
```

```python
# -------------------------------
# LIST OPERATIONS
# -------------------------------

print("=== LIST OPERATIONS ===")

numbers = [1, 2, 3]
print("Initial list:", numbers)

# Append (add at end)
numbers.append(4)
print("After append(4):", numbers)

# Insert at specific index
numbers.insert(1, 10)
print("After insert(1, 10):", numbers)

# Extend (add multiple elements)
numbers.extend([5, 6])
print("After extend([5,6]):", numbers)

# Remove element (first occurrence)
numbers.remove(10)
print("After remove(10):", numbers)

# Pop element (removes last by default)
popped = numbers.pop()
print("Popped element:", popped)
print("After pop():", numbers)

# Pop specific index
popped_index = numbers.pop(1)
print("Popped index 1:", popped_index)
print("After pop(1):", numbers)

# Access elements
print("First element:", numbers[0])

# -------------------------------
# LOOPING THROUGH LIST
# -------------------------------

print("\n=== LOOPING ===")

# For loop
print("For loop:")
for num in numbers:
    print(num)

# For loop with index
print("\nFor loop with index:")
for i in range(len(numbers)):
    print(f"Index {i} -> {numbers[i]}")

# While loop
print("\nWhile loop:")
i = 0
while i < len(numbers):
    print(numbers[i])
    i += 1

# -------------------------------
# DICTIONARY OPERATIONS
# -------------------------------

print("\n=== DICTIONARY ===")

person = {
    "name": "Surya",
    "age": 30
}

print("Initial dict:", person)

# Add new key
person["city"] = "Bangalore"
print("After adding city:", person)

# Update value
person["age"] = 31
print("After updating age:", person)

# Access value
print("Name:", person["name"])

# Loop through dict
print("\nLoop through dict:")
for key, value in person.items():
    print(key, "->", value)

# Remove key
person.pop("city")
print("After pop city:", person)

# -------------------------------
# SET OPERATIONS
# -------------------------------

print("\n=== SET ===")

my_set = {1, 2, 3}
print("Initial set:", my_set)

# Add element
my_set.add(4)
print("After add(4):", my_set)

# Remove element
my_set.remove(2)
print("After remove(2):", my_set)

# -------------------------------
# PRACTICE EXAMPLE
# -------------------------------

print("\n=== PRACTICE: EVEN NUMBERS ===")

nums = [1, 2, 3, 4, 5, 6]
even_numbers = []

for n in nums:
    if n % 2 == 0:
        even_numbers.append(n)

print("Even numbers:", even_numbers)
```
